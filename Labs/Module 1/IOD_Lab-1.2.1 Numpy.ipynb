# Numpy Exploration

````python
import numpy as np
import matplotlib.pyplot as plt
# import matplotlib

# get matplot to show charts inline
%matplotlib inline
````

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

````python
np.min(height)
````

**1. Find Dimension of Array**

Find the dimension of each array: height, weight, population
```
# Expected Ouput
np.ndim(height): 1
np.ndim(weight): 1
np.ndim(population): 2
```

````python
print("np.ndim(height): ", np.ndim(height))
print("np.ndim(weight): ", np.ndim(weight))
print("np.ndim(population): ", np.ndim(population))
````

**8. Create histograms of height and weight**

````python
#help(plt.hist)
plt.hist(height,bins=10)
plt.title('Height')
plt.grid()
plt.show()
plt.hist(weight,bins=10)
plt.title('Weight')
plt.grid()
plt.show()
````

**9. Create separate boxplots of height and weight**

````python
#help(plt.boxplot)
plt.boxplot(weight)
plt.title('Weight')
plt.grid()
plt.show()
plt.boxplot(height)
plt.title('Height')
plt.grid()
plt.show()
````

**10. Create a scatterplot of height and weight**
